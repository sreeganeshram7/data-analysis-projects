### Coding Question 4: Pandas DataFrame Filtering and Aggregation 

#### 1. Create the DataFrame with sample data

```python
import pandas as pd
df = pd.DataFrame({
'product_id': [101,102,103,104,105,106,107,108,109,110],
'product_name':['Ganesh','Ram','Moorthi','Vishnu','Sameer','bala','arun','vicky','yogi','vetri'],
'category': ['Electronics','Electronics','Clothing','Food','Food',
                 'Clothing','Electronics','Clothing','Food','Electronics'],
'Price':[800,1200,300,50,100,400,900,200,80,1500],
'quantity_sold':[30,15,50,80,75,55,60,24,22,10]
 })
```

```python
df
```

#### 2. Filter products where `price > 500` and `quantity_sold > 20`

```python
Filtered= df[(df['Price'] > 500) & (df['quantity_sold'] > 20)]
```

#### 3. Group by `category` and calculate:
   - Total revenue per category (price × quantity_sold)
   - Average price per category
   - Total quantity sold per category

```python
df['Total_Revenue']=df['Price']*df['quantity_sold'] 
```

```python
df1=df.groupby('category').agg({'Total_Revenue':'sum','Price':'mean','quantity_sold':'sum'})
```

```python
df
```

 #### 4.Find the top 3 products by total revenue

```python
top3=df.groupby('product_id')['Total_Revenue'].sum().nlargest(3).reset_index()
```

```python
top3
```

#### 5.Calculate the percentage contribution of each category to overall revenue

```python
overall=df['Total_Revenue'].sum()
```

```python
percentage=(df1['Total_Revenue']/overall)*100
```

```python
percentage.reset_index()
```

```python
# Step 6: Calculate percentage contribution of each category to overall revenue
category_revenue = df.groupby('category')['Total_Revenue'].sum()
total_revenue = category_revenue.sum()
revenue_percentage = (category_revenue / total_revenue * 100).round(2)
 
print("\n\nRevenue Contribution by Category:")
for category, percentage in revenue_percentage.items():
    revenue = category_revenue[category]
    print(f"{category}: ${revenue:,.0f} ({percentage}%)")
```
