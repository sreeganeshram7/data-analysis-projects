## Challenge 2

#### Implement z-score normalization instead of min-max.

```python
#1D array
data = np.array([10, 20, 30, 40, 50])
z_score = (data - np.mean(data)) / np.std(data)
print("Z-score Normalized Data:")
print(z_score)
```

```python
#2D array
data = np.array([
    [100, 200, 300],
    [150, 250, 350],
    [200, 300, 400]])
z_score = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
print("Z-score Normalized Data:")
print(z_score)
```

```python
z_score = (data - np.mean(data)) / np.std(data)
z_score
```

```python
#3D array
z_score = (arr - np.mean(arr, axis=(0,1))) / np.std(arr, axis=(0,1))
print(z_score)
```

```python
print("Mean:", np.mean(z_score, axis=(0,1)))
print("Std :", np.std(z_score, axis=(0,1)))
```

## Challenge 3

#### Detect and analyze seasonal patterns in quarterly data.

```python
quarterly_data = np.array([
    [120, 150, 200, 170],   # Year 1
    [130, 160, 210, 180],   # Year 2
    [125, 155, 205, 175],   # Year 3
    [135, 165, 215, 185]    # Year 4
])
quarterly_data 
```

```python
# Calculate average revenue per quarter
quarterly_mean = np.mean(quarterly_data, axis=0)
quarters=['Quarter_1','Quarter_2','Quarter_3','Quarter_4']
for q, val in zip(quarters, quarterly_mean):
    print(f"{q} Average Revenue: {val}")                 #  Quarter with highest average → peak season
                                                         #  Quarter with lowest average → weak season
```

```python
# Seasonal Index Calculation
overall_mean = np.mean(quarterly_data)
seasonal_index = quarterly_mean / overall_mean
for q, idx in zip(quarters, seasonal_index):
    print(f"{q} Seasonal Index: {idx:.2f}")             # Index > 1 → above-average quarter
                                                        # Index < 1 → below-average quarter
```

```python
# Standard Deviation
quarterly_std = np.std(quarterly_data,axis=0)
for q, sd in zip(quarters, quarterly_std):
    print(f"{q} Std Dev: {sd:.2f}")                    # Low std → consistent seasonal behaviour
                                                       # High std → unstable seasonality
```
