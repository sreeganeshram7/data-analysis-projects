```python
class EmployeeNotFoundError(Exception):
    pass
```

```python
def get_info(df,emp_id):
    try:
        if emp_id not in df['EmployeeID'].values:
            raise EmployeeNotFoundError(f"Employee {emp_id} not Found")
        emp=df[df['EmployeeID']==emp_id].iloc[0]
        return f"{emp['FirstName']} {emp['LastName']} works in {emp['Department']} as {emp['Designation']}"
    except EmployeeNotFoundError as e: 
        print(f"Custom Exception: {e}") 
```

```python
if __name__=="__main__":
    employees=load_database("employee_database(in).csv")
    if employees is not None:
        try:
            result=get_info(employees,1050)
            print(result)
        except EmployeeNotFoundError as e:
            print(e)

        try:
            result=get_info(employees,9999)
            print(result)
        except EmployeeNotFoundError as e:
            print(f"Exception Handled: {e}")
```
