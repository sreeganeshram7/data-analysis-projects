#### Q9. Fill missing values in 'Monthly_Charges' and 'Total_Charges' with the median by Contract type, and missing 'Customer_Service_Calls' with the mean.


```python
df['Monthly_Charges'] = df.groupby('Contract')['Monthly_Charges'].transform(lambda x: x.fillna(x.median()))
df['Monthly_Charges'].reset_index()
```

```python
df['Total_Charges'] = df.groupby('Contract')['Total_Charges'].transform(lambda x: x.fillna(x.median()))
df['Total_Charges'].reset_index()
```

```python
df['Customer_Service_Calls'] = df['Customer_Service_Calls'].fillna(df['Customer_Service_Calls'].mean())
df['Customer_Service_Calls'].reset_index()
```

#### Q10. Calculate the churn rate by Internet_Service type. Show the total customers, churned customers, and churn percentage for each service type.

```python
df
```

```python
df['Churn_Num'] = df['Churn'].map({'Yes': 1, 'No': 0})
churn = df.groupby('Internet_Service').agg(Total_Customers = ('Customer_ID', 'count'),Churned_Customers = ('Churn_Num', 'sum'))
churn['Churn_Percentage'] = (churn['Churned_Customers'] / churn['Total_Customers']) * 100
churn
```

#### Q11. Create new features: 'High_Monthly_Charge' (Top 25% of Monthly_Charges), 'Long_Term_Customer' (Tenure > 24 months), and 'High_Service_Calls' (Customer_Service_Calls >= 5).

```python
q75 = df['Monthly_Charges'].quantile(0.75)
df['High_Monthly_Charge'] = df['Monthly_Charges'] > q75
df['Long_Term_Customer'] = df['Tenure_Months'] > 24
df['High_Service_Calls'] = df['Customer_Service_Calls'] >= 5
df
```

#### Q14. Create a 2x2 subplot showing: (1) Churn count by Contract, (2) Average Tenure by Churn status, (3) Monthly Charges distribution by Churn, (4) Customer Service Calls vs Churn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# ---------------------- (1) Churn count by Contract ----------------------
plt.figure(figsize=(14, 12))
plt.subplot(2, 2, 1)
sns.countplot(data=df, x='Contract', hue='Churn_Num')
plt.title("Churn Count by Contract Type")
plt.xticks(rotation=20)
plt.xlabel("Contract Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# ---------------------- (2) Average Tenure by Churn ----------------------
plt.figure(figsize=(14, 12))
plt.subplot(2, 2, 2)
avg_tenure = df.groupby('Churn')['Tenure_Months'].mean().reset_index()
sns.barplot(data=avg_tenure, x='Churn', y='Tenure_Months')
plt.title("Average Tenure by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Average Tenure_Months")
plt.tight_layout()
plt.show()
```

```python
# ---------------------- (3) Monthly Charges Distribution by Churn ----------------------
plt.figure(figsize=(14, 12))
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='Churn_Num', y='Monthly_Charges')
plt.title("Monthly Charges Distribution by Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges ($)")
plt.tight_layout()
plt.show()
```

```python
# ---------------------- (4) Customer Service Calls vs Churn ----------------------
plt.figure(figsize=(14, 12))
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='Churn', y='Customer_Service_Calls')
plt.title("Customer Service Calls vs Churn")
plt.xlabel("Churn")
plt.ylabel("Customer Service Calls")

plt.tight_layout()
plt.show()
```

#### Q17. Create a 'Risk_Score' column categorizing customers: High risk (Tenure < 12 AND Monthly_Charges > 75), Medium risk (Tenure < 24 AND Monthly_Charges > 50), Low risk (others).

```python
import numpy as np
conditions = [(df['Tenure_Months'] < 12) & (df['Monthly_Charges'] > 75),(df['Tenure_Months'] < 24) & (df['Monthly_Charges'] > 50)]
choices = ['High', 'Medium']
df['Risk_Score'] = np.select(conditions, choices, default='Low')
df['Risk_Score'].reset_index()
```
